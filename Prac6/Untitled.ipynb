```python
print("============= ESD Table ==========")
print("Name \tType \tID \tAddress")
r = open("Loader.txt", "r")
line = r.readlines()
count_id = 1
add = -1
extry = {}
entry_arr = []
extrn_arr = []
start_arr = []
id_dict = {}
for i in line:
    spi = i.split(" ")
    if spi[2] == "START":
        print(" " + spi[1] + "\t " + str(count_id) + "\t " + spi[0])
        start_arr.append(spi[1])
        id_dict[spi[1]] = count_id
        count_id += 1
    elif spi[2] == "ENTRY":
        ld_symbol = spi[3]
        ld_list = ld_symbol.split(",")
        rep = ld_list[len(ld_list) - 1].replace("\n", "")
        ld_list[len(ld_list) - 1] = rep
        for k in ld_list:
            for j in line:
                spi1 = j.split(" ")
            if k == spi1[1]:
                add = spi1[0]
                extry[k] = add
                entry_arr.append(k)
                break
        print(k + " " + " LD " + "\t-1 \t " + str(add))
    elif spi[2] == "EXTRN":
        Id_symbol = spi[3]
        Id_list = ld_symbol.split(",")
        rep = ld_list[len(ld_list) - 1].replace("\n", "")
        Id_list[len(ld_list) - 1] = rep
        for m in Id_list:
            print(m + " " + " ER \t " + str(count_id) + "\t -1")
            extrn_arr.append(m)
            id_dict[m] = count_id
            count_id += 1
print("\n\n=========TXT Table=============")
print("Rel.Add Content Comment")
for i in line:
    spi2 = i.split(" ")
    if spi2[2] == "DC":
        Id_symbol = spi2[3]
        rep = ld_symbol.replace("\n", "")
        spi2[3] = rep
    sub = spi2[3][2:len(spi2[3]) - 1]
    for ey in entry_arr:
        if sub.find(ey) != -1:
            sub = sub.replace(ey, extry[ey])
    for ex in extrn_arr:
        if sub.find(ex) != -1:
            sub = sub.replace(ex, "0")
    for st in start_arr:
        if sub.find(st) != -1:
            sub = sub.replace(st, "0")
        eval1 = eval(str(sub))
        print(" " + str(spi2[0]) + "-" + str(int(spi2[0]) + 3) + " " + str(sub) + " " + sub)

print("\n=========RLD Table===========")
print("ESD_ID \tLength \tFlag \tRel.loc")
for i in line:
    spi3 = i.split(" ")
    if spi3[2] == "DC":
        ld_symbol = spi3[3]
        rep = ld_symbol.replace("\n", "")
        spi3[3] = rep
        sub = spi3[3][2:len(spi3[3]) - 1]
        for ex in extrn_arr:
            index = 0
            if sub.find(ex) != -1:
                index = sub.find(ex)
            if index == 0:
                print(" " + str(id_dict[ex]) + "\t" + "+" + "\t" + "+" + "\t" + str(spi3[0]))
            if index != 0:
                print(" " + str(id_dict[ex]) + " " + "" + sub[index - 1] + "+" + "" + str(spi3[0]))
        for st in start_arr:
            if sub.find(st) != -1:
                index = sub.find(st)
            if index == 0:
                print(" " + str(id_dict[st]) + " " + "+ " + " " + "" + " " + str(spi3[0]))
            if index != 0:
                print(" " + str(id_dict[st]) + " " + " " + " + " + " " + sub[index - 1] + " " + str(spi3[0]))
```
